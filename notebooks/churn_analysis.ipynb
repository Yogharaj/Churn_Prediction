```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data_folder = '../data/'
datasets = [f for f in os.listdir(data_folder) if f.endswith('.csv')]

for dataset in datasets:
    filepath = os.path.join(data_folder, dataset)
    df = pd.read_csv(filepath)

    print(f"\n--- EDA for {dataset} ---\n")

    print("First 5 rows of the dataset:")
    print(df.head(), "\n")
    print("Missing values in the dataset:")
    print(df.isnull().sum(), "\n")
    print("Dataset summary statistics:")
    print(df.describe(), "\n")
    if 'churn' in df.columns:
        print("Target variable distribution ('churn'):")
        print(df['churn'].value_counts())
        sns.countplot(x='churn', data=df)
        plt.title(f"Churn Distribution in {dataset}")
        plt.show()

    else:
        print("The 'churn' column is not found in this dataset.\n")
    numeric_df = df.select_dtypes(include=['number'])
    
    if not numeric_df.empty:
        plt.figure(figsize=(10, 8))
        sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
        plt.title(f"Correlation Heatmap for {dataset}")
        plt.show()
    else:
        print(f"No numeric data in {dataset} to compute correlation.\n")
```
